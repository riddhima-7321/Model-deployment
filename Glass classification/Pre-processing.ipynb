```python
df=df.drop_duplicates()
```

```python
a=df.corr()
a
```

```python
plt.figure(figsize=(40,40))
heatmap=sns.heatmap(a,annot=True,annot_kws={"fontsize": 50})
heatmap.set_xticklabels(heatmap.get_xticklabels(), fontsize=30,rotation=90)
heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=30,rotation=360)
plt.show()
```

```python
for column in df.columns:
    sns.scatterplot(x='Type',y=column,data=df)
    plt.show()
```
